## Read input data CSV

```python
import pandas as pd

#pd.set_option('display.float_format', lambda x : "{:.2e}".format(x))

df_co2_data = pd.read_csv("../datasets/owid-co2-data.csv")

#df_co2_data["gdp"].apply(lambda x : "{:e}".format(x))

df_co2_data.columns
```

```python
df_co2_france = df_co2_data[df_co2_data["country"] == "France"].copy()

df_co2_france["gdp_diff"] = df_co2_france["gdp"].diff()
df_co2_france["population_diff"] = df_co2_france["population"].diff()

df_co2_france
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [8, 9]
plt.rcParams['font.size'] = 22

plt.clf()

g = sns.barplot(y="co2", data=df_co2_france, width=0.5)

g.set_title("Annual CO2 emissions\nfrom fossil fuels in France")
g.set_ylabel("CO2 emissions (million tonnes)")
g.set_xlabel("France")

plt.savefig("../figures/emissions-france-barplot.pdf", format="pdf", bbox_inches="tight")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import re

plt.rcParams['figure.figsize'] = [8, 9]
plt.rcParams['font.size'] = 22

plt.clf()

g = sns.scatterplot(x="year", y="co2", hue="gdp_diff", data=df_co2_france, palette="viridis")

# Format the legend labels
def format_legend_label(label, pos):
    label = re.sub(r'[^\x00-\x7F]+','-', label)
    label = '{:.2e}'.format(float(label))
    return label

# Get the legend
legend = g.legend(markerscale=2)

legend.set_title('GDP Growth ($)')
# Apply the formatting function to the legend labels
for text in legend.get_texts():
    text.set_text(format_legend_label(text.get_text(), None))
#plt.legend(title='GDP')

g.set_ylabel("CO2 emissions (million tonnes)")
g.set_xlabel("Year")

plt.savefig("../figures/emissions-gdp-france.pdf", format="pdf", bbox_inches="tight")
```

```python
#corr = df_co2_france[["co2", "gdp"]].corr()

corr = df_co2_france[df_co2_france["year"] >= 1980][["co2", "gdp_diff", "population_diff"]].corr()

corr
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import re

plt.rcParams['figure.figsize'] = [8, 9]
plt.rcParams['font.size'] = 22

plt.clf()

g = sns.heatmap(corr, annot=True, cmap="coolwarm",
                xticklabels=["CO2\nEmissions", "GDP\nGrowth", "Population\nGrowth"],
                yticklabels=["CO2\nEmissions", "GDP\nGrowth", "Population\nGrowth"])

g.set_title("Pearson correlation coefficients\nof France's data starting from year 1980\n")
plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.savefig("../figures/emissions-gdp-pop-france-corr.pdf", format="pdf", bbox_inches="tight")
```

```python
df_co2_world = df_co2_data[df_co2_data["country"] == "World"].copy()
df_co2_world["gdp_diff"] = df_co2_world["gdp"].diff()
df_co2_world["population_diff"] = df_co2_world["population"].diff()

df_co2_worldfrance = pd.concat([df_co2_world, df_co2_france])

df_co2_worldfrance
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import re

plt.rcParams['figure.figsize'] = [8, 9]
plt.rcParams['font.size'] = 22

plt.clf()

g = sns.scatterplot(x="year", y="co2", hue="population_diff", style="country", data=df_co2_worldfrance, palette="plasma", s=50)



# Hack to correctly display the legend labels
dct_legend_str_labels = {"population_diff":"Population\nGrowth", "country":"Country", "World":"World", "France":"France"}

# Format the legend labels
def format_legend_label(label, pos):
    # Dealing with unusual minus symbols
    label = re.sub(r'[^\x00-\x7F]+','-', label)

    right_label = dct_legend_str_labels.get(label)
    if right_label == None:
        return '{:.2e}'.format(float(label))
    else:
        return right_label  

# Get the legend
legend = g.legend(ncol=2, fontsize=18, markerscale=2)

#g.set_yscale("log")

#legend.set_title('GDP Growth ($)')
# Apply the formatting function to the legend labels
for text in legend.get_texts():
    text.set_text(format_legend_label(text.get_text(), None))
#plt.legend(title='GDP')

g.set_ylabel("CO2 emissions (million tonnes)")
g.set_xlabel("Year")

plt.savefig("../figures/emissions-pop-france-world-flat.pdf", format="pdf", bbox_inches="tight")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import re

plt.rcParams['figure.figsize'] = [16, 9]
plt.rcParams['font.size'] = 22

plt.clf()

fig, axs = plt.subplots(1, 2)


## Plotting for France
g = sns.scatterplot(x="year", y="co2", hue="population_diff", data=df_co2_france, palette="coolwarm", s=40, ax=axs[0])

# Format the legend labels
def format_legend_label(label, pos):
    label = re.sub(r'[^\x00-\x7F]+','-', label)
    label = '{:.2e}'.format(float(label))
    return label

# Get the legend
legend = g.legend(markerscale=2)

g.set_yscale("log")

legend.set_title('Population Growth')
# Apply the formatting function to the legend labels
for text in legend.get_texts():
    text.set_text(format_legend_label(text.get_text(), None))


g.set_title("France")
g.set_ylabel("CO2 emissions (million tonnes)")
g.set_xlabel("Year")
g.set_ylim(1, 1e5)


## Plotting for World
g = sns.scatterplot(x="year", y="co2", hue="population_diff", data=df_co2_world, palette="coolwarm", s=40, ax=axs[1])

# Format the legend labels
def format_legend_label(label, pos):
    label = re.sub(r'[^\x00-\x7F]+','-', label)
    label = '{:.2e}'.format(float(label))
    return label

# Get the legend
legend = g.legend(markerscale=2)

g.set_yscale("log")

legend.set_title('Population Growth')
# Apply the formatting function to the legend labels
for text in legend.get_texts():
    text.set_text(format_legend_label(text.get_text(), None))

g.set_title("World")
g.set_ylabel("CO2 emissions (million tonnes)")
g.set_xlabel("Year")
g.set_ylim(1, 1e5)

plt.savefig("../figures/emissions-pop-france-world-logscale-subplots.pdf", format="pdf", bbox_inches="tight")
```
